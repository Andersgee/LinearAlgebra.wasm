```julia
y=zeros(3,2)
function matmul!(y, W, X, szW1, szW2)
    szX1 = szW2
    szX2 = div(length(X), szW2)
    s = 0.0
    for colX=1:szX2
        for row = 1:szW1
            s = 0.0
            for col = 1:szW2
                iW = (col - 1) * szW1 + row
                s += W[iW] * X[col + (colX-1)*szX1]
            end
            y[row + (colX-1)*szW1] = s
        end
    end
    return nothing
end
matmul!(y,W,X,3,4); y
```

```julia
y=zeros(3)
function mul!(y, W, x)
    szW1 = length(y)
    szW2 = length(x)
    s = 0.0
    for row = 1:szW1
        s = 0.0
        for col = 1:szW2
            iW = (col - 1) * szW1 + row
            s += W[iW] * x[col]
        end
        y[row] = s
    end
    return nothing
end
mul!(y,W,x); y
```

```julia
function muladd!(y, W, x, b)
    szW1 = length(y)
    szW2 = length(x)
    s = 0.0
    for row = 1:szW1
        s = 0.0
        for col = 1:szW2
            iW = (col - 1) * szW1 + row
            s += W[iW] * x[col]
        end
        y[row] = s + b[row]
    end
    return nothing
end
```
